# THIS PROJECT INCLUDES DATA EXPLORATION AND CLEANING OF COVID DATA COLLECTED FROM LINK GIVEN BELOW FOR FURTHER VISUALISATION ON TABLEAU PUBLIC

```python
import pandas as pd
pd.set_option('display.float_format','{:2f}'.format)
df=pd.read_csv('covid death.csv')
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df[df.continent.isna()]
```

```python
df.drop(df[df.continent.isna()].index.tolist(),inplace=True)
```

```python
df.shape
```

```python
df.fillna(0,inplace=True)
```

```python
df.drop(df[df['population']==0].index.tolist(),inplace=True)
```

```python
df.drop(df[df.total_cases==0].index.tolist(),inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df=df[['date','continent','location','population','total_cases','total_deaths']]
```

```python
df['date']=pd.to_datetime(df['date'],infer_datetime_format=True)
```

```python
df['date']=df['date'].dt.strftime('%Y-%m')
```

```python
df['date']
```

```python
def death_case_per(grp):
    death=list(map(lambda x,y: (x/y)*100,grp['total_deaths'],grp['total_cases']))
    cases=list(map(lambda x,y: (x/y)*100,grp['total_cases'],grp['population']))    
    grp['Infection_Spread_Percentage']=cases
    grp['Death_Percentage_']=death
    return grp
```

# Dataframe by location, total cases,total deaths,case%,death%

```python
df2=death_case_per(df.groupby(['location'],as_index=False)[df.columns.tolist()[2:]].max())
```

```python
df2
```

```python
df2.to_excel('death_case_percentage.xlsx')
```

# #Dataframe by Continent Details

```python
df3=(df.groupby(['continent','location'],as_index=False))[['population','total_cases','total_deaths']].max()
```

```python
df3.drop('location',axis=1,inplace=True)
```

```python
df3=df3.groupby('continent',as_index=False).sum()
```

```python
death_case_per(df3)
```

```python
df3.to_excel('bycontinent.xlsx')
```

```python
df4=pd.DataFrame({'Covid Infections Count ':df3.total_cases.sum(), 'Deaths ' : df3.total_deaths.sum() },index=[0])
```

```python
df4.to_excel('Total.xlsx')
```
