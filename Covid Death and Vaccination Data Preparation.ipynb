# Covid Death Data

```python
df_covid_death=pd.read_csv('owid-covid-data.csv',usecols=['continent','location','date','total_cases','new_cases','total_deaths','new_deaths'])
df_covid_death.head()
```

```python
df_covid_death.dtypes
```

# date must be converted to date time

```python
df_covid_death['date'] = pd.to_datetime(df_covid_death['date'], infer_datetime_format=True)
```

```python
df_covid_death.dtypes
```

```python
df_covid_death['date'].dt.year.head()
```

### Here I add new column with year and month

```python
df_covid_death['date(y/m)']=df_covid_death['date'].dt.strftime('%Y-%m')
```

```python
df_covid_death['date(y/m)']
```

```python
df_covid_death.drop('date',axis=1,inplace=True)
```

```python
df_covid_death.isna().sum()
```

### now filling na values with zeroes since when total death, new death = 0 , the data shows NAN

```python
df_covid_death.fillna(0,inplace=True)
```

```python
df_covid_death.isna().sum()
```

```python
df_covid_death.shape
```

### A continet can't be a zero and the values for the data with continent as zero is ambiguos so its better to drop the data 

```python
df_covid_death.continent.unique()
```

```python
df_covid_death[df_covid_death.continent==0]
```

```python
df_covid_death[df_covid_death.continent==0].index.tolist()
df_covid_death.drop(df_covid_death[df_covid_death.continent==0].index.tolist(),inplace=True)
df_covid_death.reset_index(drop=True,inplace=True)
```

### grouping by data based on date

```python
df_covid_death_bydate=df_covid_death.groupby(df_covid_death['date(y/m)'],axis=0).sum().reset_index()
```

```python
df_covid_death_bydate.dtypes
```

```python
df_covid_death_bydate['date(y/m)'] = pd.to_datetime(df_covid_death_bydate['date(y/m)'], infer_datetime_format=True)
df_covid_death_bydate
```

```python
df_covid_death_bydate.dtypes
```

```python
df_covid_death_bydate.to_excel('covid_death_date.xlsx')
```
